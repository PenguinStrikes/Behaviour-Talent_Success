```python
%load_ext autoreload
%autoreload 2
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os

PROJ_ROOT = os.path.abspath(os.path.join(os.pardir))
sys.path.append(os.path.join(PROJ_ROOT, 'src'))
import model

%matplotlib inline
```

```python
fig, ax = plt.subplots(1, 5, figsize=(12,4), sharey=True)

for index, i in enumerate([0.01, 0.1, 0.5, 0.7, 0.9]):
    w, e, t = model.talent_luck_model(1000, 0.5, 80, 10, i, 0.6, 0.1)
    ax[index].scatter(t, e, color='grey', alpha=0.2, s=5)
    ax[index].set_xlim(0,1)
    ax[index].set_yscale('log')
sns.despine()
plt.show()
```
